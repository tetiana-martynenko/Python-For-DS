### Профіль колонок набору даних:

2. Проведемо аналіз нульової активності зареєстрованих користувачів за днями тижня. Для цього
  1. Виберіть з набору даних усі записи, де кількість зареєстрованих користувачів дорівнює 0. Групуйте ці записи за днями тижня (`dayofweek`) та обчисліть кількість нульових записів для кожного дня.
  2. Після групування, використайте графік типу "bar" для візуалізації результатів. Кожна смужка на графіку має відображати кількість записів у даних для цього дня тижня, коли не було зареєстровано жодної поїздки користувачами.

```python
zero = df[df.registered == 0]
data_for_bar = zero.groupby('dayofweek')['registered'].size()
```

```python
data_for_bar.plot.bar()
plt.title('Розподіл нульової активності користувачів по днях')
plt.xlabel('День тижня')
plt.ylabel('К-ть годин');
```

Проаналізуйте графік, що спостерігаєте?

Висновки:
Найчастіше використання продукції компанії спостерігається при вологості до 40 та Температурі за відчуттями до 15. Пікові періоди по к-ть користувачів - при підвищенні температури. Отже, підвищення температури позитивно впливає на використання продукту.

```python
col_x = 'humidity'
col_y = 'registered'
color_feature = 'atemp'
plt.scatter(df[col_x], df[col_y], alpha=0.5, c = df[color_feature])
plt.title('Розподіл вологості та кількості зареєстрованих користувачів')
plt.xlabel('Вологість')
plt.ylabel('К-ть користувачів')
plt.colorbar(label="Температура за відчуттями");
```
