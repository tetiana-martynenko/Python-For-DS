# Завдання 1

У медичному центрі проводять тест на певну хворобу. Відомо, що 1% населення страждає на цю хворобу. Тест дає позитивний результат у 99% випадків для хворих і 5% випадків для здорових людей (хибнопозитивний результат). Якщо тест показав позитивний результат, яка ймовірність того, що людина насправді хвора?

---



P(H) = 0.01 - ймовірність хвороби у населення

P(-H) = 0.99 - ймовірність відсутності даної хвороби у населення

P(T|H) = 0.99 - ймовірність, що людина хвора й тест правильний

P(T|-H) = 0.05 - ймовірність, що людина здорова, але тест хибний

P(H|T)  = ?

За формулою Баєса:

$$P(H|T) = \frac{P(T|H) ⋅ P(H)}{P(T)}$$

Далі визначаємо за повною ймовірністю, яка ймовірність, що тест позитивний загалом:

$$P(T) = P(T|H) ⋅ P(H) + P(T|-H)⋅ P(-H)$$

$$P(T) = 0.99 ⋅ 0.01 + 0.05⋅ 0.99 = 0.594$$

ТОді визначаємо ймовірність позитивного результату тесту у хворої людини:

$$P(H|T) = \frac{0.99 ⋅ 0.01}{0.0594} = 0.1667$$

Відповідь: 16.67%


# Завдання 2

У компанії 70% співробітників чесні, а 30% можуть інколи брехати. Детектор брехні правильно ідентифікує чесних співробітників у 90% випадків і неправильно класифікує їх як таких, що брехали, у 10% випадків. Якщо детектор показав, що співробітник брехав, яка ймовірність того, що він насправді брехав?

Розв'язуємо аналогічно до першої задачі. Спочатку записуємо, що відомо:

P(Ch) = 0.70 - ймовірність, що співробітник чесний

P(-Ch) = 0.30 - ймовірність, що співробітник може інколи брехати

P(T|-Ch) = 0.90 - ймовірність, що детектор визначає чесних чесними

P(T|Ch) = 0.10 - ймовірність, що детектор визначає чесних людей брехунами

P(-Ch|T) = ? ймовірність, що детектор визначає правильно брехунів брехунами

За формулою Баєса:

$$P(-Ch|T) = \frac{P(T|-Ch) ⋅ P(-Ch)}{P(T)}$$

Далі визначаємо за повною ймовірністю, яка ймовірність, що тест позитивний загалом:

$$P(T) = P(T|Ch) ⋅ P(Ch) + P(T|-Ch)⋅ P(-Ch)$$

$$P(T) = 0.1 ⋅ 0.7 + 0.9 ⋅ 0.3 = 0.34$$

ТОді визначаємо ймовірність правильної роботи детектора:

$$P(+T|L) = \frac{0.9 ⋅ 0.3}{0.34} = 0.7941$$

Відповідь: 79.41%

# Завдання 3

Відомо, що предмет можна загубити в одній з трьох кімнат: у першій кімнаті з ймовірністю 0,5, у другій — з ймовірністю 0,3, і в третій — з ймовірністю 0,2. Якщо предмет загублено в кімнаті, ймовірність знайти його під час пошуку складає 0,8 для першої кімнати, 0,6 для другої і 0,4 для третьої. Якщо предмет знайшли, яка ймовірність того, що його було загублено в першій кімнаті?

A1, A2, A3 - перша, друга та третя кімнати відповідно

P(A1) = 0.5

P(A2) = 0.3

P(A3) = 0.2

f - подія знайдення предмету

P(f|A1) = 0.8
P(f|A2) = 0.6
P(f|A3) = 0.4

P(A1|f) = ?

За формулою Баєса:

$$P(A1|f) = \frac{P(f|A1) ⋅ P(A1)}{P(f)}$$

Далі визначаємо за повною ймовірністю, яка ймовірність, предмет буде знайдено:

$$P(f) = P(f|A1) ⋅ P(A1) + P(f|A2)⋅ P(A2)+ P(f|A3)⋅ P(A3)$$

$$P(f) = 0.8 ⋅ 0.5 + 0.6 ⋅ 0.3 + 0.4 ⋅ 0.2 = 0.66$$

ТОді визначаємо ймовірність, що предмет знайшли у першій кімнаті:

$$P(A1|f) = \frac{0.8 ⋅ 0.5}{0.66} = 0.606$$

Відповідь: 60.6%

A1, A2, A3 - перша, друга та третя лінії відповідно

P(A1) = 0.5

P(A2) = 0.3

P(A3) = 0.2

f - подія дефекту

P(f|A1) = 0.02
P(f|A2) = 0.03
P(f|A3) = 0.05

P(A3|f) = ?

За формулою Баєса:

$$P(A3|f) = \frac{P(f|A3) ⋅ P(A3)}{P(f)}$$

Далі визначаємо за повною ймовірністю, яка ймовірність, що був дефект:

$$P(f) = P(f|A1) ⋅ P(A1) + P(f|A2)⋅ P(A2)+ P(f|A3)⋅ P(A3)$$

$$P(f) = 0.02 ⋅ 0.5 + 0.03 ⋅ 0.3 + 0.05 ⋅ 0.2 = 0.29$$

ТОді визначаємо ймовірність, що обрана продукція випущена лінією 3:

$$P(A1|f) = \frac{0.05 ⋅ 0.2}{0.29} = 0.0344$$

# Завдання 5

Програма для розпізнавання рукописних цифр має точність 95% для цифри "3" і 98% для інших цифр. Якщо на вхід програми подано цифру "3", а програма її не розпізнала, яка ймовірність того, що насправді була подана цифра "3"? Відомо, що в 20% випадків на вхід подається цифра "3".

A -  програма розпізнала
B -  програма не розпізнала
Th, -Th - цифра 3 та інші цифри відповідно

P(A|Th) = 0.95
P(A|-Th) = 0.98
P(B|Th) = 0.05
P(B|-Th) = 0.02

P(Th) = 0.2
P(-Th) = 0.8 (антиподія P(Th))

P(Th|B) = ?(тобто програма розпізнала 3 як іншу цифру, а не 3)


За формулою Баєса:

$$P(Th|B) = \frac{P(B|Th) ⋅ P(Th)}{P(B)}$$

Далі визначаємо за повною ймовірністю, яка ймовірність, що програма не розпізнала правильно:

$$P(B) = P(B|Th) ⋅ P(Th) + P(B|-Th)⋅ P(Th)$$

$$P(B) = 0.05 ⋅ 0.2 + 0.02 ⋅ 0.8 = 0.026$$

ТОді визначаємо ймовірність того, що програма не розпізнала цифру 3:

$$P(Th|B) = \frac{0.05 ⋅ 0.2}{0.026} = 0.3846$$

Відповідь: 38.46%
