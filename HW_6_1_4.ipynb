# Завдання 1

З колоди у 52 карти випадково витягується одна карта. Яка ймовірність того, що це туз, якщо відомо, що ця карта — пікова?

P(A) = 4 / 52 = 1 / 13  - ймовірність подій, що це туз, так як в колоді 4 тузи

P(B) = 13 / 52 = 1 / 4 - ймовірність подій, що це пікова карта

P (A ∩ B)  = P(A) * P(B) = (1 / 13 ) * ( 1 / 4) = 1 / 52

P (A|B) = P(A ∩ B) / P(B) = (1 / 52) / (1 / 4) = 1 / 13

Або простіше вирішити, так як ми знаємо, що це пікова карта, а їх всього 13 в колоді. То ймовірність, що витягнемо туза складає 1 / 13, адже піковий туз може бути лише 1.

Відповідь: 1 / 13



# Завдання 2

У групі з 30 студентів є 18 хлопців і 12 дівчат. Відомо, що 7 хлопців і 5 дівчат носять окуляри. Випадково обраний студент носить окуляри. Яка ймовірність того, що цей студент — дівчина?



Так як ми знаємо, що студент носить окуляри, то загальна вибірка скорочується до 12 (7 хлопців та 5 дівчат).
P(Z) = 5 / 12  - ймовірність, що це дівчина, яка носить окуляри.

Або ж рахуємо через формулу умовної ймовірності:

P (A|B) = P(A ∩ B) / P(B)
Тобто, що обраний студент буде дівчина, за умови що вже обрано студента в окулярах


P(B) = 12 / 30 = 6 / 15 - ймовірність, що це  студент в окулярах

P (A ∩ B) = 5 / 30 - ймовірність, що це дівчина, яка носить окуляри

P (A|B) = P(A ∩ B) / P(B) = (5 / 30) / (6/ 15) = 5 / 12


Відповідь: P(Z) = 5 / 12
