````python
from math import pi  # імпортуємо лише необхідну бібліотеку


def calcArea(r):
    """ Обчислює площу кола за заданим радіусом.
    Параметр:
    r (float): Радіус кола.
    Повертає:
    float: Площа кола.
    """   # використання docstrings

    return pi * r ** 2  # пробіли навколо оператора "*"
                        # між визначенням функцій використовують 2 пустих рядка


def calcCircumference(r):
    """ Обчислює довжину кола за заданим радіусом.
    Параметр:
    r (float): Радіус кола.
    Повертає:
    float: Довжина кола.
    """  # використання docstrings

    return 2 * pi * r   # пробіли навколо оператора "*"
                        # між визначенням функцій використовують 2 пустих рядка


radius = 5  # пробіли навколо оператора "*"
print("Area:",calcArea(radius))  #для читабельності результати розділяємо по окремих рядках
print("Circumference:",calcCircumference(radius))
````

````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': pd.Categorical(np.random.choice(products, N_ROWS)),
    'product_category': pd.Categorical(np.random.choice(categories, N_ROWS)),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})

check_size2 = df.memory_usage(index=False).sum()
effect = round(100 - (check_size2 / check_size1) * 100, 2)
print(f"Обсяг зайнятої пам'яті зменшився на {effect}%")
````

````python
def CalculateVariance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    sum_squares = 0
    for n in nl:
        sum_squares = sum_squares + n**2
    mean_squares = sum_squares / len(nl)
    return mean_squares
````

````python
# виправлений код
def calculate_variance(numbers_list):
    # Переведення суми елементів списку
    total_sum = sum(numbers_list)

    # Обчислення середнього значення
    mean = total_sum / len(numbers_list)

    # Обчислення середнього значення квадратів елементів
    squared_sum = sum(n**2 for n in numbers_list)
    mean_of_squares = squared_sum / len(numbers_list)

    # Дисперсія (середнє значення квадратів мінус квадрат середнього)
    variance = mean_of_squares - mean**2

    return variance


input_list = [75, 80, 85, 90, 95]
print(calculate_variance(input_list))
````
